```python
releases = pd.read_pickle('/Users/isdeniz/Documents/github/231008-releases.pkl')
```

```python
releases_count = releases.groupby(by=['repository_name'], as_index=False)['release_id'].count()
releases_count = releases_count.sort_values('release_id', ascending=True)
releases_count
```

```python
releases['release_candidate'] = releases.tag_name.apply(lambda x: True if 'rc' in x else (True if 'RC' in x else False))
releases.head()
```

```python
releases_stable = releases.copy(deep=True)
releases_stable = releases_stable.drop(releases_stable[releases_stable['prerelease'] == True].index)
releases_stable = releases_stable.reset_index(drop=True)
releases_stable_count = releases_stable.groupby(by=['repository_name'], as_index=False)['release_id'].count()
releases_stable_count = releases_stable_count.sort_values('release_id', ascending=True)
```

```python
plt.figure(figsize=(10,5))
plt.plot(releases_count['repository_name'], releases_count['release_id'], 'r-', label='total number of releases')
plt.plot(releases_count['repository_name'], releases_stable_count['release_id'], 'b-', label='number of stable releases')
plt.xticks(range(len(releases_count['repository_name'])), releases_count['repository_name'], rotation=90)
plt.legend(loc="upper left")
plt.xlabel("repository_name")
plt.ylabel("number of releases")
plt.show()
```
